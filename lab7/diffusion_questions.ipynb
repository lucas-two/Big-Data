### 3.1. Diffusion process
Now we implement the diffusion process. Each active node will cause other nodes in the graph to become active over time. The diffusion rule is that a node gets active if at least a certain percentage of its neighbors become active. The process continues until convergence (i.e. has no new node activated).

```python
def diffusion(graph, seed_set, thr=0.5):
    active = seed_set
    has_changed = False
    
    # TODO: loop through not active nodes and activate a node if more than "thr" percent of its neighbors are active
    
    # Not active nodes
    not_active = [x for x in graph.nodes if x not in active]
    
    for n in not_active:
        total_neighbours = thr * len(list(graph.neighbors(n)))
        active_neighbours = thr * len([x for x in graph.neighbors(n) if x in active])
        
        if active_neighbours >= total_neighbours:
            active.add(n)
            has_changed = True
    if has_changed:
        return diffusion(graph, active, thr) # union of two sets
    return active

diffusion(G, {0,1})
```
