# Gap Minder (Linear Regression)

```python
# Importing our libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

%matplotlib inline
```

```python
# Reading dataset
df = pd.read_csv("gapminder.csv")
```

```python
# Looking at our data
df.head()
```

## Visualising Data
* Use seaborn to visualize the data of Gapminder

```python
# Plotting as correlation heatmap
fig, ax = plt.subplots(figsize=(10,10))

sns.heatmap(df.corr(), ax=ax, annot=True, cmap="RdYlGn")

plt.show()
```

## Apply Linear Regression
* Apply linear regression with the 'fertility' feature to predict life expectancy.

```python
# Simple Linear Regression
x = df["fertility"].values.reshape(-1, 1)
y = df["life"].values.reshape(-1, 1)

# Splitting data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)

y_pred = linear_reg.predict(x_test)
```

```python
# Scoring
slr_score = metrics.r2_score(y_test, y_pred)

# Cross validation
slr_cv_results = cross_val_score(linear_reg, x, y, cv=5)
```

```python
# Visualising
plt.scatter(x_test, y_test)
plt.ylabel('Life Expectancy')
plt.xlabel('Fertility')
plt.plot(x_test, y_pred, color='red', linewidth=2)
plt.show()
```

## Apply Multiple Linear Regression
* Apply linear regression with the all features to predict life expectancy.

```python
# Multiple Linear Regression
x = df[['population','fertility','HIV','CO2','BMI_male','GDP','BMI_female','child_mortality','Region']]
y = df['life']

# Transform categorical regions data into numerical data
regions = pd.get_dummies(x['Region'], drop_first=True)

x = x.drop('Region', axis=1)
x = pd.concat([x, regions], axis=1)

# Splitting data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

multi_linear_reg = LinearRegression()
multi_linear_reg.fit(x_train, y_train)

y_pred = multi_linear_reg.predict(x_test)
```

```python
# Scoring
mlr_score = metrics.r2_score(y_test, y_pred)

# Cross validation
mlr_cv_results = cross_val_score(multi_linear_reg, x, y, cv=5)
```

## Model Scoring
* Compare the model score when using all features to one feature in previous step. (NOTE: We have already calculated these)

```python
# Score comparison - using r-squared (how close the data fitted the regression line)
print("SLR R-squared:", slr_score)
print("MLR R-squared:", mlr_score)
```

## Applying 5-fold Cross Validation
* Apply 5-fold cross-validation (for both 2 above steps) and compare your model score accuracy.
(NOTE: We have already calculated these)

```python
# 5-fold cross validation
print("SLR CV:", slr_cv_results)
print("MLR CV", mlr_cv_results)
print("")
print("SLR CV Mean:", np.mean(slr_cv_results))
print("MLR CV Mean:", np.mean(mlr_cv_results))
```

<b>Conclusion:</b> It appears our MLR model has a higher accuracy, both with its R-squared score of 87% (being how accurate it performed on a given set of test/training data), but also with its 5-fold cross validation score of 85% (being how accurate the model performed on all avalible data).
