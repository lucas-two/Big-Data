```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from numpy.polynomial import Polynomial
from sklearn.metrics import mean_squared_error


%matplotlib inline
```

```python
plt.grid(True)
plt.plot(df['Sales'], marker='o', color='blue', label='Actual')
plt.ylabel('Sales')
plt.xlabel('Period (Months)')
plt.title('Sales Across 12 Months')
plt.legend()
plt.show()
```

## Predict future demand in month 13,14,15,16 without seasonality
* Using moving average. (k=3)
* Using exponential smoothing (alpha = 0.2 and alpha = 0.8)

```python
# Moving average
df['SMA_3'] = df['Sales'].rolling(window=3).mean()
```

```python
# Displaying moving average
plt.grid(True)
plt.plot(df['Sales'], marker='o', color='blue', label='Actual')
plt.plot(df['SMA_3'], marker='o', color='green', label='3-MA')
plt.ylabel('Sales')
plt.xlabel('Period (Months)')
plt.title('Moving Average')
plt.legend()
plt.show()
```

```python
# ------- Trying to use Thanh's method of moving average ----------

def moving_average(df, k, to_period):
    moving_average_df = pd.DataFrame(columns=['Period', 'Sales'])
    num_df = len(df)
    for m in range(0, to_period):
        if m < k:
            sale_predict = float('nan')
        else:
            history = df['Sales'][m-k:m]
            sale_predict = history.mean()
        moving_average_df.loc[m] = [m+1, sale_predict]
    moving_average_df['Period'].astype(int)
    
    return moving_average_df

# Calculate the moving average
ma_df = moving_average(df, 4, 16)

print(ma_df)

# Slice our 13 - 16 month sales into a list
ma_predictions = ma_df['Sales'].values.tolist()
ma_predictions = ma_predictions[12:16]
```

```python
# Exponeitial Smoothing
fit1 = SimpleExpSmoothing(df['Sales']).fit(smoothing_level=0.2,optimized=False)
fit2 = SimpleExpSmoothing(df['Sales']).fit(smoothing_level=0.8,optimized=False)
```

```python
# Displaying exponential smoothing
plt.plot(df['Sales'], color='blue', marker='o', label='Actual sales')
fit1.fittedvalues.plot(marker='o',  color='green', label='α = 0.2')
fit2.fittedvalues.plot(marker='o',  color='red', label='α = 0.8')

plt.grid(True)
plt.ylabel('Sales')
plt.xlabel('Preiod (Months)')
plt.title('Exponential Smoothing')
plt.legend()

plt.show()
```

## Predict future demand in month 13,14,15,16 with seasonality
* Divide the time-series into 3 cycles (Month 1-4, 5-8, 9-12), then...
 * Compute the average sales for each cycle.
 * Use linear regression on the average sales of cycles to predict the average sale for next cycle.

```python
df = df.drop(['SMA_3'], axis=1) # Remove this column as it confuses steps
```

```python
# Dividing up the data
# NOTE: This may not be the most efficent way to do this but the other methods I tried did not seem to work.
for i in range(0, 4):
    df.loc[i, 'Cycle'] = 1
    
for i in range(4,8):
    df.loc[i, 'Cycle'] = 2
    
for i in range(8,12):
    df.loc[i, 'Cycle'] = 3
```

```python
# Calculating Mean Average per Cycle
mean_I = df.loc[df['Cycle'] == 1].mean(axis=0).Sales
mean_II = df.loc[df['Cycle'] == 2].mean(axis=0).Sales
mean_III = df.loc[df['Cycle'] == 3].mean(axis=0).Sales

# Printing results
print('Average Sales for Cycle I: {0}'.format(mean_I))
print('Average Sales for Cycle II: {0}'.format(mean_II))
print('Average Sales for Cycle III: {0}'.format(mean_III))
```

```python
# Merging into an array
avg_arr = np.array([mean_I, mean_II, mean_III])
season = np.array([1,2,3])

# Fitting into a linear regression model
p = Polynomial.fit(season, avg_arr, 2)
```

```python
# Plotting the predicted line
plt.plot(*p.linspace())

plt.grid(True)
plt.ylabel('Sales')
plt.xlabel('Cycle')
plt.title('Exponential Smoothing')

plt.show()
```
