# Wine Data Exercise

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.decomposition import PCA as sklearnPCA
```

```python
df = pd.read_csv("wine.data.csv")
```

```python
label = df["Label"]

# Remove 1st column
df = df.drop('Label', axis=1)
```

## Looking at Correlation

```python
pd.plotting.scatter_matrix(df, alpha=0.2, figsize=(10,10))
plt.show()
```

<b>Concolusion:</b>
The scatter plot matrix appears to show lots of clusters in the data. They all appear to look slightly different (some look more linear like the Flavonius and Total pheonols scatters). From the histograms it looks like the data isn that normally distributed.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), ax=ax, annot=True)
plt.show()
```

 <b>Can you find any pairs of attributes which have large correlation?</b> It appears that the variables 'Total phenols' and 'Flavanoids' have the highest correlation. It also appears that 'Flavanoids' and 'OO280' have the second highest correlation of 0.79.

## Normalising The Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)
x_scaled_array = scaler.transform(df)
normalized_data = pd.DataFrame(x_scaled_array, columns=df.columns)

normalized_data.head()
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans

k_means_clustering = KMeans(n_clusters=3, random_state=0)

res = k_means_clustering.fit(normalized_data) # first we fit
label_KM = k_means_clustering.predict(normalized_data) + 1 # then we predict (note: we are adding 1 so its 1, 2, 3 not 0, 1, 2)

label_KM
```

```python
normalized_data["cluster"] = label_KM.astype("float64")
sns.pairplot(normalized_data,hue="cluster", diag_kind="hist")
plt.show()
```

## PCA
PCA allows us to perform dimension reduction. We want our 13 variables to become 2D so we can visualise it better.

```python
from sklearn.decomposition import PCA

pca_data = PCA(n_components=2)
principalComponents_data = pca_data.fit_transform(normalized_data)
```

```python
k_means_clustering_pca = KMeans(n_clusters=3, random_state=0)

res_pca = k_means_clustering_pca.fit(principalComponents_data) # first we fit
label_KM_pca = k_means_clustering_pca.predict(principalComponents_data) + 1

label_KM_pca
```

```python
from sklearn.metrics.cluster import adjusted_rand_score

# original label vs K-Means label
adjusted_rand_score(label, label_KM)
```

```python
# original label vs PCA label
adjusted_rand_score(label, label_KM_pca)
```

<b>Conclusion:</b> It appears both clusters are moderately similar to the original labels.
